```
import pandas as pd

url = 'https://raw.githubusercontent.com/dm-fedorov/pandas_basic/master/data/football.csv'

df = pd.read_csv(url)
df
```

```
l1=df.groupby(['Position'])[['Wage','Value']].mean()
l1.sort_values(['Value'],ascending=False)
```

```
l1=df.groupby(['Club'])['Wage'].agg(['mean','median'])
l2=l1.loc[l1['mean']==l1['median']]
len(l2.index)
```

```
l1.loc[l1['mean']==l1.max().values[0]]
```

```
df.groupby(['Club'])['Wage'].sum()['Chelsea']
```

```
df.loc[(df['Nationality']=='Argentina') & (df['Age']==30)]['Wage'].max()
```

```
s=df.loc[df['Nationality']=='Argentina']
s=s.loc[s['Club']=='FC Barcelona']
l1=s['Strength'].max()
l2=s['Balance'].max()
print(str(l1)+ ', '+str(l2))
```
